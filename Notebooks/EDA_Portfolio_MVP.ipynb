## MOTORISTA DE APLICATIVO EM SÃO GONÇALO/RJ - MVP

## 1. INTRODUÇÃO

    O conjunto de dados analisados, foi uma construção de acordo com os conhecimentos sobre Banco de Dados relacionais adquiridos na gradução tecnológica de Ciência de Dados. O período analisado é de Agosto a Dezembro de 2024. A base de dados foi pensada de acordo com os conceitos de normalização do banco de dados, e podem ser acessados por aqui : [(https://github.com/renanpacheco2/Portfolio-Uber/blob/main/data/Uber_Renan_Portfolio.xlsx)]

    🎯 Objetivo do Projeto

    O objetivo deste projeto foi analisar as corridas realizadas no aplicativo Uber para:

    1. Avaliar os valores praticados nas corridas e identificar padrões de precificação.

    2. Entender a tendência de ganhos por hora, avaliando a eficiência e rentabilidade ao longo do tempo.

    3. Analisar a relação entre valor recebido e quilometragem percorrida (R$/km).

    4. Identificar as localidades mais lucrativas com base no volume e valor médio das corridas.

    5. Mapear horários de pico e períodos de maior movimentação.

## 2. Modelo Conceitual (Entidades e Relacionamentos)

```
                   ┌─────────────────────┐
                   │     Motoristas      │
                   │  id_motorista (PK)  │
                   │  nome, idade, etc.  │
                   └──────────┬──────────┘
                              │
                              │ (FK)
                              │

┌───────────────┐     ┌───────▼─────────┐      ┌─────────────────────────┐
│  Aplicativos  │     │    Corridas     │      │         Carros          │
│ id_app (PK)   │◄────┤ id_viagem (PK)  │─────►│ id_carro (PK)           │
│ nome_app      │     │ id_motorista(FK)│      │ modelo, ano, marca      │
│               │     │ id_local(FK)    │      │ rendimento, combustível │
└───────────────┘     │ id_pagamento(FK)│      └─────────────────────────┘
                      │ id_carro(FK)    │
                      │ id_app(FK)      │
                      │ id_valor(FK)    │
                      │ data, duração   │
                      │ dia_da_semana   │
                      │ horário         │
                      └───────┬─────────┘
                              │
                              │

               ┌──────────────▼──────────────┐
               │         Localidade          │
               │      id_local (PK)          │
               │   bairro, cidade, estado    │
               └─────────────────────────────┘


               ┌──────────────▼──────────────┐
               │        Valor_Viagem         │
               │      id_viagem (PK)         │
               │ pg_motorista, turbo+        │
               │ dinâmico, valor_total       │
               └─────────────────────────────┘


               ┌──────────────▼──────────────┐
               │     Tipo de Pagamento       │
               │    id_pagamento (PK)        │
               │         método              │
               └─────────────────────────────┘
```


## 2.1) Benefícios do modelo relacional
    - Integridade e redução de redundância.
    - Flexibilidade para expansão do projeto.
    - Boa representação da origem transacional dos dados.

## Evolução proposta
Como o foco do projeto também é analítico, uma evolução futura é migrar para Star Schema para simplificar agregações, melhorar performance e tornar o modelo mais aderente a BI e analytics.
